```python
def makeSelectors(soup):
    selectors = {}
    for styles in soup.select('style'):
        css = cssutils.parseString(styles.encode_contents())
        for rule in css:
            if rule.type == rule.STYLE_RULE:
                style = rule.selectorText
                selectors[style] = {}
                for item in rule.style:
                    propertyname = item.name
                    value = item.value
                    selectors[style][propertyname] = value
    return selectors
```

```python
#selectors
#soup.find('div', class_="pc1").find_all('div')

def getCSSClass(page_class, css_element):
    element_list = []
    all_class = soup.find('div', class_=page_class).find_all('div', class_=True)
    for classes in all_class:
        class_list = classes['class']
        for e_class in class_list:
            if css_element in e_class:
                element_list.append(e_class)
    return list(set(element_list))
```

```python
def getCSSValue(selectors, element_list, property_name):
    value_list = {}
    for element in element_list:
        value_list[element] = selectors['.{}'.format(element)][property_name]
    return value_list
```

```python
def getElementByClass(element_list, class_list):
    match_element = []
    for element in element_list:
        element_class = element['class']
        if(all(x in element_class for x in class_list)):
            match_element.append(element)
    return match_element
```

```python
def getClassByValue(class_list, value_name):
    match_class = []
    for e_class in class_list:
        for value in value_name:
            if value == e_class[0:len(value)]:
                match_class.append(e_class)
    return match_class
```

```python
def getDivByWord(page_class, word):
    word_div = []
    div_list = soup.find('div', page_class)
    for div in div_list:
        text = div.text
        text_no_space = text.replace(" ", "")
        if word == text_no_space:
            word_div.append(div)
    if len(word_div) != 0:
        return word_div
    else:
        return False
```

```python
def getBrowser(file_name):
    html_file = os.path.join(os.getcwd(), 'pdf/html', file_name)
    browser.get("file:///" + html_file)
    return browser
```

```python
def getPCList(browser):
    page_class_list = []
    for e in browser.find_element_by_id('page-container').find_elements_by_xpath('*'):
        page_id = e.get_attribute('id')
        page_class = page_id.replace('pf', 'pc')
        page_class_list.append(page_class)
    return page_class_list
```

```python
def getDiv(browser, page_class_list, bottom_limit=(930.0, 80.0)):
    div_list = []
    for page_class in page_class_list:
        for e in browser.find_element_by_class_name(page_class).find_elements_by_xpath('.//div'):
            # Remove Page Number
            bottom_px = e.value_of_css_property('bottom')
            bottom = float(bottom_px.replace('px', ''))
            if ((bottom < bottom_limit[0]) & (bottom > bottom_limit[1])):
                div_list.append(e)
                #print(e.get_attribute("textContent"))
        time.sleep(1)
    return div_list
```

```python
def getTypeOfDiv(browser, div_list):
    try:
        ele_judge = browser.find_element_by_xpath("//*[contains(text(), '판')]")
    except NoSuchElementException:
        ele_judge = None
    try:
        ele_decision = browser.find_element_by_xpath("//*[contains(text(), '주')]")
    except NoSuchElementException:
        ele_decision = None
    
    if ((ele_judge is not None) & (ele_decision is not None)):
        head_ff = ele_judge.value_of_css_property('font-family')
        head_fs1 = ele_judge.value_of_css_property('font-size')
        head_fs2 = ele_judge.value_of_css_property('font-size')
    else:
        return None
    
    ele_judge_idx = div_list.index(ele_judge)
    ele_decision_idx = div_list.index(ele_decision)
    
    ele_division = div_list[ele_judge_idx+1]
    division_ff = ele_division.value_of_css_property('font-family')
    division_fs = ele_division.value_of_css_property('font-size')
    
    div_type_list = []
    for idx, div in enumerate(div_list):
        div_ff = div.value_of_css_property('font-family')
        div_fs = div.value_of_css_property('font-size')
        if div_ff == head_ff:
            div_type_list.append((div, 'head'))
        else:
            if ((div_ff == division_ff) & (float(div_fs.replace('px', '')) < float(division_fs.replace('px', ''))+1.0) & (float(div_fs.replace('px', ''))+1.0 > float(division_fs.replace('px', '')))):
                if (ele_judge_idx < idx) & (ele_decision_idx > idx):
                    div_type_list.append((div, 'division'))
                else:
                    div_type_list.append((div, 'body'))
    return div_type_list
```

```python
browser = getBrowser('21056.html')
page_class_list = getPCList(browser)
div_list = getDiv(browser, page_class_list)
div_type_list = getTypeOfDiv(browser, div_list)
```
